```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("larger_sales_dataset.csv")

display(df.head(15))
print(df.describe)
print(df.info)

```

```python
df.isnull().sum()

df = df.dropna()
display(df)
```

```python
df.duplicated().sum()

df = df.drop_duplicates()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
display(df)

```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
display(df)
```

```python
df['Revenue'] = df['Total Price']
df['Order Value Bucket'] = pd.cut(
    df['Revenue'],
    bins=[0, 100, 500, 1000, np.inf],
    labels=['Low', 'Medium', 'High', 'Very High']
)
display(df)
display(df)
```

```python
df['Order Status'].value_counts()
```

```python
df = df[~df['Order Status'].isin(['Completed', 'Cancelled', 'Refunded'])]

display(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
```

```python
total_revenue = df['Revenue'].sum()
print("Total Revenue:", total_revenue)
```

```python
monthly_revenue = df.groupby('Month')['Revenue'].sum().reset_index()

plt.figure()
plt.plot(monthly_revenue['Month'].astype(str), monthly_revenue['Revenue'])
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
category_revenue = df.groupby('Product Category')['Revenue'].sum().sort_values(ascending=False)

plt.figure()
category_revenue.plot(kind='bar')
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
category_quantity = df.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False)

plt.figure()
category_quantity.plot(kind='bar')
plt.title("Quantity Sold by Category")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.show()
```

```python
status_counts = df['Order Status'].value_counts()

plt.figure()
status_counts.plot(kind='bar')
plt.title("Order Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```python
cancelled_revenue = df[df['Order Status'] == 'Cancelled']['Revenue'].sum()
print("Revenue Lost due to Cancellation:", cancelled_revenue)
```

```python
total_customers = df['Customer ID'].nunique()
print("Total Customers:", total_customers)
```

```python
orders_per_customer = df.groupby('Customer ID')['Order ID'].count()

plt.figure()
orders_per_customer.hist()
plt.title("Orders per Customer Distribution")
plt.xlabel("Number of Orders")
plt.ylabel("Customers")
plt.show()
```

```python
repeat_customers = (orders_per_customer > 1).sum()
one_time_customers = (orders_per_customer == 1).sum()

print("Repeat Customers:", repeat_customers)
print("One-time Customers:", one_time_customers)
```

```python
payment_revenue = df.groupby('Payment Type')['Revenue'].sum().sort_values(ascending=False)

plt.figure()
payment_revenue.plot(kind='bar')
plt.title("Revenue by Payment Type")
plt.xlabel("Payment Type")
plt.ylabel("Revenue")
plt.show()
```

```python
total_orders = df['Order ID'].nunique()
AOV = total_revenue / total_orders

print("Average Order Value (AOV):", AOV)
```

```python
df.to_csv("cleaned_sales_data.csv", index=False)
```

```python
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())
```

```python
df['Month'] = df['Month'].astype(str)
query = """
SELECT SUM(Revenue) AS total_revenue
FROM df
"""
pysqldf(query)
```

```python
query = """
SELECT Month, SUM(Revenue) AS revenue
FROM df
GROUP BY Month
ORDER BY Month
"""
pysqldf(query)
```

```python
query = """
SELECT `Product Category`, SUM(Revenue) AS revenue
FROM df
GROUP BY `Product Category`
ORDER BY revenue DESC
"""
pysqldf(query)
```

```python
query = """
SELECT
    (SUM(CASE WHEN `Order Status` = 'Cancelled' THEN 1 ELSE 0 END) * 100.0 / COUNT(*))
    AS cancellation_rate
FROM df
"""
pysqldf(query)
```

```python
query = """
SELECT COUNT(*) AS repeat_customers
FROM (
    SELECT `Customer ID`
    FROM df
    GROUP BY `Customer ID`
    HAVING COUNT(`Order ID`) > 1
)
"""
pysqldf(query)
```
